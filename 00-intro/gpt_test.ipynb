```python
prompt = """
 업비트와 코인원의 대표적인 차이가 뭐야?
 왜 사람들이 대부분 업비트를 쓰고 그에 비해 코인원은 잘 안쓰지?
 그저 후발주자라서?
 """

from openai import OpenAI
client = OpenAI(api_key='OPENAI_API_KEY')

response = client.responses.create(
    model = 'gpt-4.1-nano',
    input = prompt
    )

print(response.output_text)
```
