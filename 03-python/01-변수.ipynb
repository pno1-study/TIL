### f-string (f'')
String 안에 변수를 녹여서 써야할 때 사용

```python
name = '마이클'
age = 24
# greeting = name + '님 환영합니다' + name + ' 님의 등급은 일반 입니다.' + name + '님의 나이는 age살 입니다.'
# f-string (f'')
greeting = f'{name}님 환영합니다.\n{name}님의 나이는 {age}살 입니다.'
print(greeting)
```

```python
# String 안에서 다옴표 사용하기
title = 'lily\'s bag'
story = '그는 말했다 \"안녕?\" 이라고.'
print(title, story)
```
