### Short-circuit Evaluation (단축 평가)
- 앞에서 값이 확실시 되면, 그 뒤는 확인하지 않는다.
- 조건 판별시 뒷부분을 확인하지 않아도 됨.

```python
a = True and True and True and False and True and False
# ?   ->  ?   ->   ?   ->   ! [종료, 뒤에 볼 필요 없음]
b = False or False or True or False
# ?   ->   ?   ->   ! [종료, 뒤에 볼 필요 없음]
print (a, b)
```

```python
x = 1 and 2 and 3 and 0 and 4
y = 0 or None or '' or [] or [None] or 10 or 'asdf'
print (x, y)

x = '0211'
# x가 값이 있으면, x를 사용
# x가 값이 없으면(''),
x or '기본값' # '0211'
x = ''
x or '기본값' # '기본값'
```

```python
# 초를 시간, 분, 초 로 바꾸기
# 계산을 통해 h, m, s 에 시간, 분, 초로 전환
seconds = 3675  # 1시간 1분 15초
h = 3675//3600
m = 3675%3600//60
s = 3675%3600%60
f'{h}시간 {m}분 {s}초'
```

```python
# 세자리 수 분리하기
# 숫자를 백, 십, 일의 자리로 분리하여 변수에 저장하기
n = 472
a = n//100
b = (n-a*100)//10
c = n-(a*100)-(b*10)
print(a, b, c)
```
