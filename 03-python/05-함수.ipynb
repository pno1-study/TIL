````python
# name: plus_one / in: 숫자 / out: 숫자 +1
def plus_one(num):
    return num+1
# 함수 호출(call)
plus_one(1)
````

````python
# n번 인사하는 함수
# name: hell_ntimes

def hello_ntimes(n):
    for x in range(n):
        print('반가워')
hello_ntimes(3)
````

````python
# 3제곱 해주는 함수
# name: cube, in: 숫자 1개, out: 숫자의 세제곱
def cube(n):
    return n**3
cube(2)
````

````python
# 원기둥의 부피를 구하는 함수
# name: cylinder, in: 반지름/높이, out: 부피(pi*r^2*h)
def cylinder(n, m):
    return 3.14*n**2*m
cylinder(2, 2)
````

````python
# 리스트 내부의 숫자들의 평균을 구해주는 함수
# name: get_avg, in: 숫자들로 만들어진 리스트, out: 평균
def get_avg(n):
    return sum(n)/len(n)
get_avg([1, 2, 3, 4, 5])
````

````python
# 비밀번호 체크해주는 함수 (숫자 포함, 8글자 이상)
# name: check_pw, in: 비밀번호 문자열, out: True/False

def check_pw(password):
    is_ok = False
    if len(password) >= 8:
        for char in password:
            if char in '0123456789':
                is_ok = True
    return is_ok

check_pw('abcd1234')  # True
check_pw('abcdefg')  # False
````

````python
# BMI 기준으로, 비만여부 확인해주는 함수 (BMI = 체중(kg) / 키(m)^2)
# 남성 27 이상일 경우 비만
# 여성 25 이상일 경우 비만

def check_bmi(n, m, l):
    m = m/100
    bmi = n/(m**2)
    print(bmi)
    if l == 'M' and bmi >= 27:
        print('비만')
    elif l == 'M' and bmi < 27:
        print('정상')
    elif l == 'F' and bmi >= 25:
        print('비만')
    else :
        print('정상')

# 몸무게, 키, M -> male
check_bmi(100, 175, 'M') # 예상값 = 비만
# 몸무게, 키, F -> female
check_bmi(50, 160, 'F') # 예상값 = 정상
check_bmi(67, 173.5, 'M') # 예상값 = 비만
````
