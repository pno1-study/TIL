```python
# Person
# 키, 몸무게, 나이, 이름
# 인사, 공부

# python에서 클래스 이름은 PascalCase로 짓는다. my_good_friend / MyGoodFriend
class Person:
    #1. 특성 (attribute)
    height = 165
    weight = 55
    age = 20
    name = '인간'
    #2. 행동 (method)
    def greeting(self):
        return f'안녕하세요, {self.name} 입니다. {self.age}살 입니다.'
    def study(self):
        self.iq += 5
        
```

```python
# Person 개념적 대상을 -> 실제 생성
p1 = Person()
p2 = Person()
```

```python
# 특성
p1.age = 30
p1.name = '김로봇'
p1.weight = 80
p1.height = 230
# 행동
p1.greeting()
```

```python
p2.age = 23
p2.name = '박사람'
p2.weight = 45
p2.height = 155
p2.greeting()
```
