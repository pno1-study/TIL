```python
# 문제 1. 점수 평균과 합격 판별
# 학생 점수 리스트가 있다.
scores = [80, 75, 90, 60, 100]
# 평균을 구하시오.
score_sum = 0
for i in scores:
    score_sum += i
score_avg = score_sum/len(scores)
# 평균이 80 이상이면 "합격" 아니면 "불합격" 출력
if score_avg >= 80:
    print('합격')
else:
    print('불합격')
```

```python
# 문제 2. 최대값 찾기
# 숫자 리스트가 있다.
# 가장 큰 값을 찾아 출력하시오.
# (힌트: max() 없이 직접 비교해보기)
numbers = [3, 7, 2, 9, 5]
number_max = 0
for i in numbers:
    if number_max < i:
        number_max = i
    else:
        pass
print(number_max)
```

```python
# 문제 3. 짝수만 저장하기
# 짝수만 새로운 리스트에 담아서 출력하시오.
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
number_even = []
for i in numbers:
    if i%2 == 0:
        number_even.append(i)
print(number_even)
```

```python
# 문제 4. 이름 존재 여부 확인
# 변수 target 에 저장된 이름이 리스트 안에 있으면
# "존재합니다" 아니면 "없습니다" 출력
target = "민수"
names = ["철수", "영희", "민수", "지영"]
target_num = len(names)
list_out = 0
for i in names:
    if target == i:
        target_num = list_out
        list_out += 1
    else:
        list_out += 1
if target_num != list_out:
    print('존재합니다.')
else:
    print('없습니다.')
```

```python
# 문제 5. 딕셔너리 값 접근
student = {
    "name": "철수",
    "age": 20,
    "score": 85
}
# 점수가 80 이상이면 "우수 학생" 출력
# 아니면 "보통 학생" 출력
if student['score'] >= 80:
    print('우수 학생')
else:
    print('보통학생')
```

```python
# 문제 6. 재고 확인 시스템
inventory = {
    "apple": 10,
    "banana": 0,
    "orange": 5
}
# 각 과일의 재고가
# 0이면 → "품절"
# 1 이상이면 → "판매 가능" # 출력하시오.
for i in inventory:
    if inventory[i] == 0:
        print(f'{i} = 품절')
    else:
        print(f'{i} = 판매 가능')
```

```python
# 문제 7. 리스트 안 딕셔너리 평균 구하기
students = [
    {"name": "철수", "score": 80},
    {"name": "영희", "score": 90},
    {"name": "민수", "score": 70}
]
# 전체 평균 점수 구하기
# 평균 이상인 학생 이름 출력
score_sum = 0
for i in students:
    score_sum += i['score']
score_mean = score_sum/len(students)
students_meanup = []
for i in students:
    if i['score'] >= score_mean:
        students_meanup.append(i['name'])
print(students_meanup)
```

```python
# 문제 8. 등급 매기기
#점수 리스트:
scores = [95, 82, 67, 74, 88]
# 각 점수에 대해
# 90 이상 → A
# 80 이상 → B
# 70 이상 → C
# 그 외 → D
# 등급 리스트를 새로 만들어 출력하시오.
classes = []
for i in scores:
    if i >= 90:
        classes.append('A')
    elif i >= 80:
        classes.append('B')
    elif i >= 70:
        classes.append('C')
    else:
        classes.append('D')
print(classes)
```

```python
# 문제 9. 장바구니 총 금액 계산
cart = [
    {"item": "노트북", "price": 1000000, "qty": 1},
    {"item": "마우스", "price": 20000, "qty": 2}
]
# 총 결제 금액 계산
# 100만원 이상이면 "고액 구매자" 출력
pay_total = 0
for i in cart:
    pay_total += i['price'] * i['qty']
if pay_total >= 1e6:
    print('고액 구매자')
```

```python
# 문제 10. 비밀번호 검사기
password = "abc123"
# 조건:
# 길이가 8자 이상이고
# 숫자가 포함되어 있으면 "사용 가능"
# 아니면 "사용 불가" 출력하시오.
# (힌트: len(), in  사용. for 문으로 한글자씩 체크)
password_number=[]
for i in password:
    if i in str(list(range(0, 10))):
        password_number.append(i)
if len(password) >= 8 and len(password_number) >= 1:
    print('사용 가능')
else:
    print('사용 불가')
```
